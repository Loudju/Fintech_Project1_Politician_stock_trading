```python
import requests
import json
import pandas as pd
import hvplot.pandas
import quiverquant
from datetime import datetime
import yfinance as yf
import numpy as np
```

```python
quiver = quiverquant.quiver("c28eacfdf71675aa0dcad7feaa75eedfff51c547")
senate_party_df = quiver.senate_trading()
```

```python
 def senate_trading(ticker=""):
        if len(ticker)>0:
            url = "https://api.quiverquant.com/beta/historical/senatetrading/"+ticker
        else:
            url = "https://api.quiverquant.com/beta/live/senatetrading"
        
        headers = {'accept': 'application/json',
'X-CSRFToken': 'TyTJwjuEC7VV7mOqZ622haRaaUr0x0Ng4nrwSRFKQs7vdoBcJlK9qjAS69ghzhFu',
'Authorization': 'Token c28eacfdf71675aa0dcad7feaa75eedfff51c547'}
        r = requests.get(url, headers=headers)
        j = json.loads(r.content)
        df = pd.DataFrame(j)
        df["Date"] = pd.to_datetime(df["Date"])
        return df  
```

```python

senate_dem = senate_party_df[senate_party_df["Party"]=="D"]
senate_dem_aapl = senate_dem[senate_dem["Ticker"]=="AAPL"]
```

```python
aapl = yf.Ticker("AAPL")
aapl_hist = aapl.history(start="2020-01-01", end="2022-06-03")
display(aapl_hist)
```

```python
senate_aapl = senate_trading("AAPL")

senate_aapl = senate_aapl[(senate_aapl['Date'] > '2020-01-01') & (senate_aapl['Date'] <= '2022-06-03')]

display(senate_dem_aapl.groupby(['Date', 'Senator','Transaction']).sum())
```

```python
senate_dem_aapl_close = senate_dem_aapl.merge(aapl_hist, left_on='Date', right_on='Date')
senate_dem_aapl_close.set_index('Date')
senate_dem_aapl_close['Amount'] = senate_dem_aapl_close['Amount'].astype(float)
senate_dem_aapl_close = senate_dem_aapl_close.drop(columns=['Range','Open','High','Low','Volume','Dividends','Stock Splits'])
display(senate_dem_aapl_close)
```

```python
print(senate_dem_aapl_close.groupby('Transaction').get_group('Purchase'))
print(senate_dem_aapl_close.groupby('Transaction').get_group('Sale'))
```

```python
appl_purchase_dem_df=(senate_dem_aapl_close.groupby('Transaction').get_group('Purchase'))
appl_purchase_dem_df['Shares']= (appl_purchase_dem_df['Amount'] / appl_purchase_dem_df['Close'])
appl_purchase_dem_df = (appl_purchase_dem_df.round(2))
display(appl_purchase_dem_df)


```

```python
appl_sale_dem_df=(senate_dem_aapl_close.groupby('Transaction').get_group('Sale'))
appl_sale_dem_df['Shares']= (appl_sale_dem_df['Amount'] / appl_sale_dem_df['Close'])
appl_sale_dem_df = (appl_sale_dem_df.round(2))
display(appl_sale_dem_df)
```

```python
democrat_aapl_df= pd.concat([appl_sale_dem_df,appl_purchase_dem_df])
ron_wyden_aapl_df= (democrat_aapl_df.groupby('Senator').get_group("Wyden, Ron"))
ron_wyden_aapl_df

```

```python
aapl_hist.hvplot(x='Date', y='Close', height=500, width=1000)*ron_wyden_aapl_df.hvplot.scatter(x='Date', y='Close', color='Transaction').opts(size=10)
```
