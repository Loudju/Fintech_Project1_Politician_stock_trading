```python
def senate_trading_def(ticker=""):
        if len(ticker)>0:
            url = "https://api.quiverquant.com/beta/historical/senatetrading/"+ticker
        else:
            url = "https://api.quiverquant.com/beta/live/senatetrading"
        
        headers = {'accept': 'application/json',
'X-CSRFToken': 'TyTJwjuEC7VV7mOqZ622haRaaUr0x0Ng4nrwSRFKQs7vdoBcJlK9qjAS69ghzhFu',
'Authorization': 'Token c28eacfdf71675aa0dcad7feaa75eedfff51c547'}
        r = requests.get(url, headers=headers)
        j = json.loads(r.content)
        df = pd.DataFrame.from_dict(j)
        return df
```
